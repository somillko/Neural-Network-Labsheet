```python
df_x = pd.DataFrame(boston.data,columns = boston.feature_names) #independent variable yani ki x
df_y = pd.DataFrame(boston.target) #dependent variable yani ki y
```

```python
df_x.head(13)
```

```python
df_y.head(10)
```

```python
regr = linear_model.LinearRegression()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_x,df_y,test_size = 0.2,random_state = 4)
```

```python
x_train.head()
```

```python
y_test.values.dtype
```

```python
regr.fit(x_train,y_train)
```

```python
regr.intercept_
```

```python
print("regression coefficient :\n",regr.coef_)
print("MSE %0.2f" % np.mean((regr.predict(x_test)-y_test)**2))
print("variance score %0.2f" % regr.score(x_test,y_test))
```
